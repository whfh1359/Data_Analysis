### merge 를 통해 전체 확진일자 만들기
* 이미지 출처 및 Pandas 공식문서 보기 : [How to combine data from multiple tables? — pandas documentation](https://pandas.pydata.org/pandas-docs/stable/getting_started/intro_tutorials/08_combine_dataframes.html#join-tables-using-a-common-identifier)

<img src="https://pandas.pydata.org/pandas-docs/stable/_images/08_merge_left.svg">

```python
# 확진자가 없는 날도 일자에 표현이 되도록 전체 일자와 확진 데이터를 merge 로 합쳐줍니다.

# df_days.merge?
#df_days.merge(df_daily_case) #error가 난다. key값으로 삼을만한 게 없다. 
#가장 좋은 error해결 방법은 shift+tab키 누르고 right_on과 left_on 등 key 값 삼아줘야 한다.
df_days.merge(df_daily_case, left_on = "확진일자", right_on = df_daily_case.index)

#근데 전부 다 나올줄 알아는데 안나온다. 왜냐면 how는 default 값이 inner join이다.
#모든 확진일자를 다 나오게 하고 싶으면 how = left(확진일자 데이터프레임) 으로 해준다.
df_days.merge(df_daily_case, left_on = "확진일자", right_on = df_daily_case.index, how = "left")
all_day = df_days.merge(df_daily_case, left_on = "확진일자", right_on = df_daily_case.index, how = "left")
all_day.head()
```

## 누적 확진자 수 구하기

```python
# 확진수를 fillna를 통해 결측치를 0으로 채워주고 누적해서 더해줍니다.

all_day["확진수"] #결측치가 있고 없는지 나온다.
all_day["확진수"].sum() #1321명이 누적확진수였다.
all_day["확진수"].head(10) #보면 1+3+3+1 으로 해도 되고 nan 무시해도 되는데 0으로 채워준다.
all_day["확진수"].fillna(0).cumsum()
all_day["누적확진"] = all_day["확진수"].fillna(0).cumsum() #cumsum은 누적합
all_day
```

```python
# 연도를 제외하고 월-일로 "일자" 컬럼 만들기

# 연도를 제외하고 일자만 가지고 온다.

all_day["확진일자"].astype(str).map(lambda x: x[-5:]) #string으로 일단 만들고 slicing 할려 한다.
all_day["일자"] = all_day["확진일자"].astype(str).map(lambda x: x[-5:]) #string으로 일단 만들고 slicing 할려 한다.
all_day
# 일자가 유일한 값이므로 index로 만들어도 될 것 같다.
all_day.head()
```

* [    ] : 일치시킬 문자 세트의 패턴
* [가나다] : 가 or 나 or 다 중에 하나를 포함하고 있는지
* [가-힣] : 한글 가부터 힣까의 문자 중 하나를 포함하고 있는지
* [0-9] : 0~9까지의 숫자 중 하나를 포함하고 있는지
* [^0-9] : 숫자를 포함하고 있지 않음
* [^가-힣] : 한글이 포함되어 있지 않음
* [가-힣+] : 한글이 하나 이상 포함되는지


```python
import re #regular expression 약자
# 정규표현식 라이브러리를 불러옵니다.
# 숫자외의 데이터는 제거하는 정규표현식
# #7265 접촉(추정)

re.sub("[0-9]", "", "7265 접촉(추정)") #0에서 0까지 공백을 숫자로 바꿔라
re.sub("[^0-9]", "", "7265 접촉(추정)") #숫자를 제외한 나머지 문자를 공백으로 바꿔라

# 이걸 가지고 함수 만들어 보겠다.
def get_number(text):
    return re.sub("[^0-9]", "", text)

get_number("#7265 접촉(추정)")
```

```python
# 함수를 통해 숫자외의 문자를 제거하는 get_number 함수를 만듭니다.

# 위의 공백->숫자, 공백-> 문자 등을 가지고 함수 만들어 보겠다.
def get_number(text):
    return re.sub("[^0-9]", "", text)
```

### 구와 월별 해외 확진자 수
* groupby, crosstab, pivot_table 로 같은 결과의 테이블이 나오도록 구합니다.
#### groupby 로 빈도수 구하기

```python
df_oversea.groupby(["지역", "월"])["연번"].count() # ["월, 지역"] 으로 하면 순서가 바뀐다.
month_gu = df_oversea.groupby(["월", "지역"])["연번"].count().unstack()
month_gu = month_gu.fillna(0).astype(int)
month_gu.style.background_gradient(cmap = "Greens")
month_gu.T.style.background_gradient(cmap = "Greens") # Transpose 해서 볼 수도 있다.
```

#### crosstab 으로 빈도수 구하기

```python
# pd.crosstab? 
pd.crosstab(df_oversea["월"], df_oversea["지역"]) 
month_gu = pd.crosstab(df_oversea["월"], df_oversea["지역"]) 
month_gu.style.bar() #groupby와 같은 결과를 crosstab으로 그리기. 너무 작은건 색깔 안나온다.
```
