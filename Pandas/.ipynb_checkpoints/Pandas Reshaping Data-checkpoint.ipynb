```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("mpg")
df.shape
```

```python
df.head()
```

```python
df.sort_values('mpg')
```

```python
df.sort_values('cylinders')
```

```python
df.sort_values('mpg',ascending=False).head() #ascending=false는 순차정렬이 아니라 역으로 정렬하겠다는 뜻이다.
#기본값은 ascending=True는 순차정렬
```

```python
df.rename(columns = {'model_year':'year'}) #model_year을 year로
```

```python
df = df.rename(columns = {'model_year':'year'}) #저장하고 싶으면 df를 다시 저장.
```

```python
df
```

```python
df.sort_index() #index별로 정리를 하겠다.
```

```python
df.reset_index() #index를 새로 또 만든 것.
```

```python
df.drop(columns=['mpg','year','name']) #drop을 inplace하거나 df로 다시 저장해주어야만 drop된 df가 저장이 되어 반영이 된다.
```
