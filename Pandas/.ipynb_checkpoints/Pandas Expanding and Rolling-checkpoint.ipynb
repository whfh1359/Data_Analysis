````python
import pandas as pd
import numpy as np
%matplotlib inline
````

````python
s = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2015', periods=1000))
s.plot()
````

````python
s = s.cumsum()
s.plot() # 
````

````python
r = s.rolling(window = 30) # 60 
r.mean()  #window size  .  .
````

````python
s.plot(style='k--')
````

````python
#       rolling .
s.plot(style='k--')
r.mean().plot(style='k') #window 30     .
````

````python
# series ,  dataframe  
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=pd.date_range('1/1/2000', periods=1000),
                  columns=['A', 'B', 'C', 'D'])
df.head()
````

````python
df = df.cumsum()
df.head()
````

````python
df.plot()
````

````python
df.rolling(window=60).sum().plot(subplots=True)
````

````python
#rolling window size    , expanding 
df.rolling(window=len(df), min_periods=1).mean()[:5]
````

````python
df.rolling(window=len(df), min_periods=1).mean().plot()
````

````python
df.expanding(min_periods=1).mean().plot()
````
