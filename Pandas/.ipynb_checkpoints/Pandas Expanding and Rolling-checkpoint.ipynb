````python
import pandas as pd
import numpy as np
%matplotlib inline
````

````python
s = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2015', periods=1000))
s.plot()
````

````python
s = s.cumsum()
s.plot() #누적 그래프
````

````python
r = s.rolling(window = 30) #처음에는 60으로 실습
r.mean()  #window size만큼 값을 이동시켜준다. 앞에가 비워졌다.
````

````python
s.plot(style='k--')
````

````python
#이동 평균 구할 때 추세선 구할 때 rolling을 사용한다.
s.plot(style='k--')
r.mean().plot(style='k') #window 30으로 바꾸면 추세선이 앞으로 바짝 붙는다.
````

````python
#앞에꺼는 series 실습, 이거는 dataframe 형태 실습
df = pd.DataFrame(np.random.randn(1000, 4),
                  index=pd.date_range('1/1/2000', periods=1000),
                  columns=['A', 'B', 'C', 'D'])
df.head()
````

````python
df = df.cumsum()
df.head()
````

````python
df.plot()
````

````python
df.rolling(window=60).sum().plot(subplots=True)
````

````python
#rolling은 window size 지정해서 그 값만큼 이동, expanding은 누적
df.rolling(window=len(df), min_periods=1).mean()[:5]
````

````python
df.rolling(window=len(df), min_periods=1).mean().plot()
````

````python
df.expanding(min_periods=1).mean().plot()
````
