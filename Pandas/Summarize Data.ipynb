# 8) 판다스로 apply 활용하기 lambda 익명함수 사용하기

### apply(function)

````python
df.apply(lambda x : x[0]) #익명함수를 사용할 수 있다.
````

````python
df['species'].apply(lambda x : x[:3])
````

````python
df['species_3']= df['species'].apply(lambda x : x[:3])
````

````python
df # 새로 species_3 column 생성 가능
````

````python
#함수를 만든다
def smp(x):
    #뒤에서 3번째까지의 문자를 가져오는 함수
    x = x[-3:]
    return x
````

````python
df['species-3'] = df['species'].apply(smp)
````
