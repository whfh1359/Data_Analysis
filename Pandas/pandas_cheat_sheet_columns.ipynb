# 6) 일부 컬럼을 기준으로 데이터 가져오기 Subset Variables (Columns)

```python
import pandas as pd
import seaborn as sns #seaborn은 데이터 시각화 툴이며, anaaconda에 기본적으로 설치 되어있다.
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
df = df[['sepal_width', 'sepal_length', 'species']] #이 관련된 데이터셋이 iris 데이터에 있어서 seaborn을 불러와야 한다.
df.head()
```

```python
columns = ['sepal_width', 'sepal_length', 'species']
df[columns]
```

```python
columns = ['sepal_width', 'sepal_length', 'species']
df['sepal_width'] #하나의 column만 가져오고 싶을 때는 특정 column명만 넣기
```

```python
df['sepal_width']
```

```python
df.sepal_width #단 한글인 경우에는 안될 가능성이 있으므로 감싸서 불러오기.
```

```python
df.filter(regex = 'regex') #regex는 정규표현식을 의미한다. 조건이 없으므로 현재는 index만 가져왔다.
```

```python
df.filter(regex = '\.') #정규표현식에 조건을 달면 '.'이 들어간 column을 가져온다.
```

```python
df.filter(regex = '_') # '_'만 들어간 것 뽑아온다.
```

```python
df.filter(regex = 'length$' )
```

```python
df.filter(regex = '^sepal') # 특정 문자.대소문자 맞추기
```

```python
df.filter(regex = '^x[1-5]$') #'^x[1-5]$' 이런 컬럼이 없으므로 불러오지 않는다. x로시작하고 1-5까지 가져오기
```

```python
df.filter(regex = 'es$') #es로 끝나는거
```

```python
df.filter(regex = '^(?!species$).*') #not을 의미하므로 species 빼고 보여줘
```

```python
df.loc[2:5,'sepal_width':'petal_width'] #위에 올려서 df petal column있는걸로 실행 후 이 part 실행하기
#앞에는 행(2:5). 뒤에는 index(column명)으로 indexing 한다고 보면 된다.
```

```python
df.iloc[:, [1,2,4]] #loc는 에러난다. iloc로 해야한다. loc는 끝까지 다 가져오는데 iloc는 끝-1까지 가져온다.
```

```python
df.iloc[:, [1,3]]
```

```python
df.iloc[-5: , [1,3]] #뒤에서 5개까지만 보고싶다.
```

```python
df.loc[df['sepal_length'] > 3, ['sepal_length','sepal_width']] # 3보다 큰 값 불러오기/ 앞에 꺼 행, 뒤에는 열 
```
