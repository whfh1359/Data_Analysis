# 12) melt, pivot 으로 Tidy Data 만들기

```python
import pandas as pd
```

```python
pd.melt?
```

```python
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                   'B': {0: 1, 1: 3, 2: 5},
                   'C': {0: 2, 1: 4, 2: 6}})
df
```

```python
pd.melt(df, id_vars=['A'], value_vars=['B']) #A는 id로, B는 value로 row로 이동된것을 볼 수 있따.
```

```python






pd.melt(df, id_vars=['A'], value_vars=['B','C']) #value에 들어간 것을 볼 수 있다.
```

```python
pd.melt(df,value_vars=['A','B','C']) 
```

```python
#df의 Method Chaining part에서 rename 한다.
```

```python
pd.melt(df,value_vars=['A','B','C']).rename(columns={
'variable' : 'var',
'value' : 'val'}) #column값 변경된 것 적용된 것 볼 수 있다.
```

```python
df.pivot?
```

```python
df2 = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
...                            'two'],
                    'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                    'baz': [1, 2, 3, 4, 5, 6]})
df2
```

```python
df2.pivot(index='foo', columns='bar', values='baz')
```

```python
#method chaining 응용
df3 = df2.pivot(index='foo', columns='bar', values='baz').reset_index() #index 새로 만듦.
df3
```

```python
df3.melt?
```

```python
df3.melt(id_vars=['foo'], value_vars=['A','B', 'C']).sort_values('bar')
```

```python
df3.melt(id_vars=['foo'], value_vars=['A','B', 'C']).sort_values(['bar','foo'])
#두개로도 가능하다.
```

```python
df3.melt(
    id_vars=['foo'], 
    value_vars=['A','B', 'C']).sort_values(
    ['foo','bar']).rename(columns = {'value':'baz'})
#pivot으로 했다가 다시 melt로 되돌려보는 것 + method chaining(rename, reset_index) 한 것
```
