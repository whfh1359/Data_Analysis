# 신한카드 -> 삼성카드 변환 

```python
def shinhan_to_samsung(x,y):
    x = x.replace(", ", "|")
    shinhan_df_species.loc[shinhan_df_species["업종"].str.contains(x), "업종"] = y
    return 
```

```python
shinhan_df_species = shinhan_df[shinhan_df["업종"].notnull()].copy()
#shinhan_df_species.head(40)


category_1 = "M027_가전/가구, M020_인테리어"
shinhan_to_samsung(category_1, "가전/가구/인테리어")

category_2 = "M019_서비스"
shinhan_to_samsung(category_2, "가정생활/서비스")

category_3 = "M021_학원, M022_유아교육, M023_교육용품"
shinhan_to_samsung(category_3, "교육/학원")

category_4 = "M017_미용서비스, M018_화장품"
shinhan_to_samsung(category_4, "미용")

category_5 = "M006_백화점"
shinhan_to_samsung(category_5, "백화점/상품권/아울렛")

category_6 = "M013_스포츠/문화/레저, M014_스포츠/문화/레저용품"
shinhan_to_samsung(category_6, "스포츠/문화/레저")

category_7 = "M015_숙박, M016_여행"
shinhan_to_samsung(category_7, "여행/교통")

category_8 = "M001_한식, M002_일식/중식/양식, M003_제과/커피/패스트푸드, M004_기타요식, M005_유흥"
shinhan_to_samsung(category_8, "요식/유흥")

category_9 = "M024_병원, M025_약국, M026_기타의료"
shinhan_to_samsung(category_9, "의료")

category_10 = "M028_자동차판매, M029_자동차서비스/용품"
shinhan_to_samsung(category_10, "자동차")

category_11 = "M030_주유"
shinhan_to_samsung(category_11, "주유")

category_12 = "M011_의복/의류, M012_패션/잡화"
shinhan_to_samsung(category_12, "패션/잡화")

category_13 = "M008_편의점"
shinhan_to_samsung(category_13, "편의점")

category_14 = "M007_할인점/슈퍼마켓, M009_기타유통, M010_음/식료품"
shinhan_to_samsung(category_14, "할인점/마트")

shinhan_df_species.head()
shinhan_df_species[shinhan_df_species["카드이용건수(천건)"].isnull()]
```
