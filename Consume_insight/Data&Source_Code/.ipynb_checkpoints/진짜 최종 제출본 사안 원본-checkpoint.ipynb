# 오프라인 데이터 전처리 

```python
samsung_df = pd.read_csv("Samsungcard.csv", encoding = 'CP949')
shinhan_df = pd.read_excel("Shinhancard.xlsx", encoding = 'CP949')
```

```python
#삼성 카드 데이터 전처리
#소비일자 datetime으로 변경하여 월, 주 뽑아내기
samsung_df['소비일자'] = samsung_df['소비일자'].astype(str)
samsung_df['소비일자'] = pd.to_datetime(samsung_df['소비일자'])

#연도-월 column생성
samsung_df["년도"] = samsung_df["소비일자"].astype(str).map(lambda x : x[-10:-6])
samsung_df["월"] = samsung_df["소비일자"].dt.month

#주 column 생성
samsung_df["주1"] = samsung_df[(samsung_df["년도"] == "2020")]["소비일자"].dt.week + 52   
samsung_df["주2"] = samsung_df[(samsung_df["년도"] == "2019")]["소비일자"].dt.week
samsung_df["주1"] = samsung_df["주1"].fillna(0)
samsung_df["주2"] = samsung_df["주2"].fillna(0)
samsung_df["주"] = samsung_df["주1"] + samsung_df["주2"]
samsung_df["주"] = samsung_df["주"].astype(int)

del samsung_df["주1"]
del samsung_df["주2"]
```

```python
#신한 카드 데이터 전처리
#소비일자 datetime으로 변경하여 월, 주 뽑아내기
shinhan_df= shinhan_df[['일별', '성별', '연령대별', '업종', '카드이용건수(천건)']]
shinhan_df['소비일자'] = shinhan_df['일별'].astype(str)
shinhan_df['소비일자'] = pd.to_datetime(shinhan_df['소비일자'])
shinhan_df["년도"] = shinhan_df["소비일자"].astype(str).map(lambda x : x[-10:-6])
shinhan_df["월"] = shinhan_df["소비일자"].dt.month

#주 column 생성
shinhan_df["주1"] = shinhan_df[(shinhan_df["년도"] == "2020")]["소비일자"].dt.week + 52   
shinhan_df["주2"] = shinhan_df[(shinhan_df["년도"] == "2019")]["소비일자"].dt.week
shinhan_df["주1"] = shinhan_df["주1"].fillna(0)
shinhan_df["주2"] = shinhan_df["주2"].fillna(0)
shinhan_df["주"] = shinhan_df["주1"] + shinhan_df["주2"]
shinhan_df["주"] = shinhan_df["주"].astype(int)

del shinhan_df["주1"]
del shinhan_df["주2"]
samsung_df.rename(columns = {'소비업종' :'업종'}, inplace = True)
```

# 온라인 데이터 전처리

```python
# 엑셀 테이터 파일명 리스트 생성
files=glob.glob('상품 카테고리 데이터_KDX 시각화 경진대회 Only/[!K]*.xlsx')
# 카테고리 데이터 불러와서 리스트로 저장
category_data_all = [pd.read_excel(filename) for filename in files]
online_all = category_data_all[0]
for i in range(1, len(category_data_all)):
    online_all = pd.concat([online_all, category_data_all[i]])
```

# 온라인 데이터 카테고리를 쿠팡 기준으로 병합

```python
#os유형과 구매금액은 인사이트 도출에 무의미함에 삭제했다.
online_all=online_all.drop(['OS유형','구매금액'],axis=1)
online_all.rename(columns = {'구매날짜' : '소비일자','카테고리명':'업종','고객성별':'성별','구매수':'소비건수','고객나이':'연령대'}, inplace = True)
category_all = online_all['업종'].unique()
```

```python
# 대분류 생성 함수 정의
def online_to_coupang(category_list,big_category):
    for i in category_list:
        online_all.loc[online_all['업종'].str.contains(category_all[i]), '업종'] = big_category
# 대분류 번호 지정
digital=[0,1,14,6,58,20,46,63,41,22,23,28,56,57,47,42]
baby=[43,44,54]
beauty=[61,32,18]
food=[2,5,15,45]
kitchen=[53,52,51,50]
fashion=[3,13,34,35,37,39,60]
sports=[7,9,11,12,17,30,31,33,49,62,4,59]
toy=[8,19,55,40]
book=[16]
pet=[21]
life=[24,26,36,38,27,29]
car=[48]
# 대분류 생성
online_to_coupang(digital, '가전/가구/홈인테리어')
online_to_coupang(baby, '출산/유아동')
online_to_coupang(beauty, '미용')
online_to_coupang(food, '식품')
online_to_coupang(kitchen, '주방용품')
online_to_coupang(fashion, '패션의류/잡화')
online_to_coupang(sports, '스포츠/레저')
online_to_coupang(toy, '완구/취미')
online_to_coupang(book, '도서/음반/DVD')
online_to_coupang(pet, '반려동물용품')
online_to_coupang(life, '생활용품')
online_to_coupang(car, '자동차용품')
```

```python
#datetime 바꾸기
online_all['소비일자']=online_all['소비일자'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
#년도와 주를 따로 column으로 만들고 2020년도에 해당하는 경우 주차에 52를 더해준다.
#2019년 12월 30,31일은 1주차로 인식이 되기 때문에 52를 따로 더해준다.
#2020년 6월 29,30일은 79주차로 2일밖에 없기에 시각화에 적합하지 않다. 
online_all['년도']=online_all['소비일자'].dt.year
online_all['월']=online_all['소비일자'].dt.month
online_all['주']=online_all['소비일자'].dt.week
online_all.loc[online_all['년도']==2020,'주']+=52
online_all.loc[online_all['소비일자'].isin(['2019-12-30','2019-12-31']),'주']+=52
online_all_79=online_all[online_all['주']==79].index
online_all=online_all.drop(online_all_79)
```

```python
online_all=online_all[['소비일자','년도','월','주','업종','성별','연령대','소비건수']]
```

# 온라인 업종별 소비건수 시각화

```python
online_all_2019=online_all[online_all['년도']==2019]
online_all_2020=online_all[online_all['년도']==2020]
online_species_year_count_2019=online_all_2019.groupby('업종')["소비건수"].sum().to_frame().sort_values(by='소비건수',ascending = False)
online_species_year_count_2020=online_all_2020.groupby("업종")["소비건수"].sum().to_frame().sort_values(by='소비건수',ascending = False)
```

```python
#온라인 코로나 전 업종별 소비량 시각화
online_species_year_count_2019=(online_species_year_count_2019/2).sort_values(by='소비건수',ascending  = False)
online_species_year_count_2019.plot.barh(figsize=(20,6),title='온라인 코로나 전 업종별 소비량')
plt.xlim(0,3.5*1e7)
```

```python
#온라인 코로나 전 업종별 소비량 시각화
online_species_year_count_2020.plot.barh(figsize=(20,6),title='온라인 코로나 후 업종별 소비량')
plt.xlim(0,3.5*1e7)
```

# 오프라인 업종별 소비건수 시각화

```python
offline_all_2019=offline_all[offline_all['년도']=='2019']
offline_all_2020=offline_all[offline_all['년도']=='2020']
offline_species_year_count_2019=offline_all_2019.groupby('업종')["소비건수"].sum().to_frame().sort_values(by='소비건수',ascending = False)
offline_species_year_count_2020=offline_all_2020.groupby("업종")["소비건수"].sum().to_frame().sort_values(by='소비건수',ascending = False)
```

```python
#오프라인 코로나 전 업종별 소비량 시각화
offline_species_year_count_2019=(offline_species_year_count_2019/2).sort_values(by='소비건수',ascending  = False)
offline_species_year_count_2019.plot.barh(figsize=(20,6),title='오프라인 코로나 전 업종별 소비량')
plt.xlim(0,3.5*1e9)
```

```python
#오프라인 코로나 후 업종별 소비량 시각화
offline_species_year_count_2020.plot.barh(figsize=(20,6),title='오프라인 코로나 후 업종별 소비량')
plt.xlim(0,3.5*1e9)
```

전체적으로 2019년대비 2020년의 업종별 소비가 줄었다는 것을 알 수 있다. 이를 통해 코로나19의 영향으로 소비가 줄었다고 시사할 수 있으며
업종별 소비와 코로나의 상관관계를 보아 각 업종별로 소비의 트렌드가 변화를 보고 변화가 있는 업종의 목록을 추출한다.
그 후 변화한 업종들의 소비층을 분석하고 포스트 코로나시대에 맞춘 마켓팅의 방향을 본다.

# 코로나 시대 기준 설립

```python
# 크롤링으로 가져올 url 주소를 입력합니다.
url = "http://www.seoul.go.kr/coronaV/coronaStatus.do"
# 판다스의 read_html 을 통해 위 url에 있는 불러옵니다.
table = pd.read_html(url)
covid = table[3]
covid["확진일자"] = pd.to_datetime("2020-"+ covid["확진일"].str.replace(".", "-"))
covid['주']=covid['확진일자'].dt.week
covid['월']=covid['확진일자'].dt.month
```

```python
#월에 따라서 코로나 확진자수를 살펴본 후 포스트 코로나 시대의 시작점을 찾아본다.
covid['월'].value_counts().sort_index().plot.bar()
```

```python
#주에 따라 확진자수를 살펴봤는데 최근 확진자수가 많아서 초반에 급증한 시점을 보기 힘들어 33주부터의 데이터는 잘라서 본다.
covid['주'].value_counts().sort_index().plot.bar(rot=0,figsize=(15,4))
```

```python
#이 결과 8주 후부터 급증함을 알 수 있기에 포스트 코로나의 기준을 8주 이후로 잡는다.
covid[covid['주']<33]['주'].value_counts().sort_index().plot.bar(rot=0,figsize=(15,4))
```

```python
covid['주'].value_counts().sort_index().plot.bar(rot=0,figsize=(15,4))
plt.axvline(4.5,color='red',label='포스트 코로나 시작')
plt.axvspan(4.5,46,color='red',alpha=0.2)
plt.legend(['포스트 코로나 시작'],loc='upper right')
```

# 코로나 전 업종별 나이, 성별 소비순위 

```python
#오프라인 데이터에 60대 데이터까지 존재하기에 10대부터 60대로 구분하였다.
online_all = online_all[(online_all["연령대"] > 0) & (online_all["연령대"] < 70)]
```

```python
#코로나 전 후 온라인/오프라인 데이터 구분
before_corona_online=online_all[online_all['주']<61]
after_corona_online=online_all[online_all['주']>60]
```

```python
#온라인 그래프 그리는 함수
def online_draw_1(df):
    plt.axvline(df.mean(), color = 'red')
    plt.axvline(df.mean(), color = 'red')
    plt.axvspan(df.mean(), df.max(), color = 'red', alpha = 0.2)
    plt.legend(["업종별 소비건수 평균 이상"], loc = 'upper right')
    plt.xlim(0,0.6*1e7)
```

```python
before_online_df_baby= before_corona_online[before_corona_online["업종"] == '출산/유아동']
before_online_df_beauty= before_corona_online[before_corona_online["업종"] == '미용']
before_online_df_kitchen= before_corona_online[before_corona_online["업종"] == '주방용품']
before_online_df_sports= before_corona_online[before_corona_online["업종"] == '스포츠/레저']
before_online_df_toy= before_corona_online[before_corona_online["업종"] == '완구/취미']
```
