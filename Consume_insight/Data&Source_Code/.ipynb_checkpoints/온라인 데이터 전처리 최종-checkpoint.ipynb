```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
files=glob.glob('상품 카테고리 데이터_KDX 시각화 경진대회 Only/[!K]*.xlsx')
```

```python
category_data_all = [pd.read_excel(filename) for filename in files]
```

```python
digital=[1,14,6,58,20,46,63,41,22,23,28,56,57,47,42]
baby=[44,54]
beauty=[32,18]
food=[5,15,45]
kitchen=[52,51,50]
fashion = [13,34,35,37,39,60]
sports = [9,11,12,17,30,31,33,49,62,4,59]
toy = [19,55,40]
life = [26,36,38,27,29]
tour = 25
```

```python
#출산/유아동
online_baby=category_data_all[43]
for i in baby:
    online_baby=pd.concat([online_baby,category_data_all[i]])
#가전/가구/홈인테리어    
online_digital=category_data_all[0]
for i in digital:
    online_digital=pd.concat([online_digital,category_data_all[i]])
#미용    
online_beauty=category_data_all[61]
for i in beauty:
    online_beauty=pd.concat([online_beauty,category_data_all[i]])
#식품        
online_food=category_data_all[2]
for i in food:
    online_food=pd.concat([online_food,category_data_all[i]])
#주방용품        
online_kitchen=category_data_all[53]
for i in kitchen:
    online_kitchen=pd.concat([online_kitchen,category_data_all[i]])
#패션의류/잡화
online_fashion = category_data_all[3]
for i in fashion:
    online_fashion = pd.concat([online_fashion, category_data_all[i]])

#스포츠/레저
online_sports = category_data_all[7]
for i in sports:
    online_sports = pd.concat([online_sports, category_data_all[i]])

#완구/취미
online_toy = category_data_all[8]
for i in toy:
    online_toy = pd.concat([online_toy, category_data_all[i]])

#도서/음반/DVD
online_book = category_data_all[16]

#반려동물용품
online_pet = category_data_all[21]

#생활용품
online_life = category_data_all[24]
for i in life:
    online_life = pd.concat([online_life, category_data_all[i]])

#자동차용품
online_car = category_data_all[48]

#여행/티켓
online_tour = category_data_all[10]
online_tour = pd.concat([online_tour, category_data_all[25]])
```

```python
online_list=[online_digital,online_baby,online_beauty,online_kitchen,online_food,
             online_fashion, online_sports, online_toy, online_book, online_pet, 
             online_life, online_car, online_tour]
```

```python
for i in online_list:
    i['구매날짜']=i['구매날짜'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
```

```python
#년도와 주를 따로 column으로 만들고 2020년도에 해당하는 경우 주차에 52를 더해준다.
for i in online_list:
    i['년도']=i['구매날짜'].dt.year
    i['주']=i['구매날짜'].dt.week
    i.loc[i['년도']==2020,'주']+=52
```

```python
#os유형과 구매금액은 인사이트 도출에 큰 영향을 끼치지 않다고 판단하여 drop시킨다.
def drop_col(df):
    return df.drop(columns=['OS유형','구매금액'],inplace=True)
for i in online_list:
    i = drop_col(i)
```

```python
for i in online_list:
    i['월']=i['구매날짜'].dt.month
```
