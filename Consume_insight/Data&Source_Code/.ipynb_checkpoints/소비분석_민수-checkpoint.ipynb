# 신한카드 -> 삼성카드 변환 

```python
def shinhan_to_samsung(x,y):
    x = x.replace(", ", "|")
    shinhan_df_species.loc[shinhan_df_species["업종"].str.contains(x), "업종"] = y
    return 
```

```python
shinhan_df_species = shinhan_df[shinhan_df["업종"].notnull()].copy()
#shinhan_df_species.head(40)


category_1 = "M027_가전/가구, M020_인테리어"
shinhan_to_samsung(category_1, "가전/가구/인테리어")

category_2 = "M019_서비스"
shinhan_to_samsung(category_2, "가정생활/서비스")

category_3 = "M021_학원, M022_유아교육, M023_교육용품"
shinhan_to_samsung(category_3, "교육/학원")

category_4 = "M017_미용서비스, M018_화장품"
shinhan_to_samsung(category_4, "미용")

category_5 = "M006_백화점"
shinhan_to_samsung(category_5, "백화점/상품권/아울렛")

category_6 = "M013_스포츠/문화/레저, M014_스포츠/문화/레저용품"
shinhan_to_samsung(category_6, "스포츠/문화/레저")

category_7 = "M015_숙박, M016_여행"
shinhan_to_samsung(category_7, "여행/교통")

category_8 = "M001_한식, M002_일식/중식/양식, M003_제과/커피/패스트푸드, M004_기타요식, M005_유흥"
shinhan_to_samsung(category_8, "요식/유흥")

category_9 = "M024_병원, M025_약국, M026_기타의료"
shinhan_to_samsung(category_9, "의료")

category_10 = "M028_자동차판매, M029_자동차서비스/용품"
shinhan_to_samsung(category_10, "자동차")

category_11 = "M030_주유"
shinhan_to_samsung(category_11, "주유")

category_12 = "M011_의복/의류, M012_패션/잡화"
shinhan_to_samsung(category_12, "패션/잡화")

category_13 = "M008_편의점"
shinhan_to_samsung(category_13, "편의점")

category_14 = "M007_할인점/슈퍼마켓, M009_기타유통, M010_음/식료품"
shinhan_to_samsung(category_14, "할인점/마트")

shinhan_df_species.head()
shinhan_df_species[shinhan_df_species["카드이용건수(천건)"].isnull()]
```

# shinhan card 데이터를 samsung 카드 데이터와 column 이름, 순서 동일하게 만들기

```python
samsung_df.rename(columns = {'소비업종' :'업종'}, inplace = True)
samsung_df
#samsung_df[samsung_df["소비건수"].isnull()]
```

```python
shinhan_df_species['카드이용건수(천건)'] = shinhan_df_species['카드이용건수(천건)'] * 1000
shinhan_df_species['카드이용건수(천건)'] = shinhan_df_species['카드이용건수(천건)'].astype(int)
shinhan_df_species
```

```python
shinhan_df_species.rename(columns = {'연령대별' : '연령대', '카드이용건수(천건)' : '소비건수'}, inplace = True)
shinhan_df_species = shinhan_df_species[['소비일자', '업종', '성별', '연령대', '소비건수','연도', '월', '주']]
shinhan_df_species["성별"].head(7)
shinhan_df_species.loc[shinhan_df_species["성별"] == "F","성별"] = "여성"
shinhan_df_species.loc[shinhan_df_species["성별"] == "M","성별"] = "남성"
shinhan_df_species
#shinhan_df_species[shinhan_df_species["소비건수"].isnull()]
```

```python
samsung_df.loc[samsung_df.업종 == "가전/가구", "업종"] = "가전/가구/인테리어" 
samsung_df.head()
```

```python
offline_df_2019_1 = offline_df[((offline_df["연도"] == "2019") &(offline_df["월"] < 7) )]   
offline_df_2019_2 = offline_df[((offline_df["연도"] == "2019") &(offline_df["월"] > 6) )]   
offline_df_2020_1 = offline_df[((offline_df["연도"] == "2020") &(offline_df["월"] < 7) )]

offline_df_2020_1[offline_df_2020_1["주"] == 79]
```

```python
species_year_count_1 = offline_df_2019_1.groupby(["연도", "업종", "월"])["소비건수"].sum()
species_year_count_1 = species_year_count_1.to_frame()
species_year_count_2 = offline_df_2019_2.groupby(["연도", "업종", "월"])["소비건수"].sum()
species_year_count_2 = species_year_count_2.to_frame()
species_year_count_3 = offline_df_2020_1.groupby(["연도", "업종", "월"])["소비건수"].sum()
species_year_count_3 = species_year_count_3.to_frame()

offline_df_2019_2[offline_df_2019_2["소비일자"] == '2019-12-30']
offline_df_2019_2["소비일자"] = "2019-12-30"
```
